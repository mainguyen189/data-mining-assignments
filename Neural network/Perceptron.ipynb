```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import Perceptron

from itertools import product
```

```python
#logic gate
# Inputs 
data = [[0, 0], [0,1], [1,0], [1,1]]
#Labels for AND
labelsAND = [0, 0, 0, 1]
#Labels for OR
labelsOR = [0, 1, 1, 1]
#Labels for XOR
labelsXOR = [0, 1, 1, 0]
labels = [labelsAND,labelsOR,labelsXOR] 
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1,3,1)
plt.scatter([point[0] for point in data], [point[1] for point in data], c = labelsAND)
plt.title('AND points')
plt.subplot(1,3,2)
plt.scatter([point[0] for point in data], [point[1] for point in data], c = labelsOR)
plt.title('OR points')
plt.subplot(1,3,3)
plt.scatter([point[0] for point in data], [point[1] for point in data], c = labelsXOR)
plt.title('XOR points')
plt.show()
```
