```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("E:\\newing\\bank-additional-full.csv",sep=';')
data.head()
```

```python
cat_cols_df = data.select_dtypes(include=['object'])
cat_cols = list(cat_cols_df.columns)
print(cat_cols)

#code all categorical columns
for name in cat_cols:
    col = pd.Categorical(data[name])
    data[name] = col.codes
    
```

```python
data_features = data.drop(['y'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, precision_recall_curve, classification_report, confusion_matrix, roc_curve, roc_auc_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_features,data['y'],test_size=0.3)
```

```python
classifiers = {'RdFrst': RandomForestClassifier(n_estimators=100), 
               'LgRg': LogisticRegression(multi_class='ovr',solver = 'liblinear')}

n_classifiers = len(classifiers)

for index, (name, classifier) in enumerate(classifiers.items()):
    classifier.fit(x_train,np.ravel(y_train))

    y_predict = classifier.predict(x_test)
    accuracy = accuracy_score(y_test, y_predict)
    print(classification_report(y_test, y_predict))
    print(y_predict)
    print(accuracy_score(y_test, y_predict))
    print(confusion_matrix(y_test, y_predict))
```
